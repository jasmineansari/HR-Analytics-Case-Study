```python
#enumerate function
for i,j in enumerate(ftr):
    print(i,j)
```

```python
fig = plt.subplots(figsize=(20,25))
for p,q in enumerate(ftr):
    plt.subplot(3,2,p+1)
    plt.subplots_adjust(hspace=0.5)
    sns.countplot(x = q, data = data, hue = "left")
    plt.xticks(rotation=90)
```
